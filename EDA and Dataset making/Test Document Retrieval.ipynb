```python
from datasets import load_dataset

train_dataset = load_dataset("SkelterLabsInc/JaQuAD", split='train')

test_dataset = load_dataset("SkelterLabsInc/JaQuAD", split='validation')
```

```python
import pandas as pd
test_data = pd.DataFrame()
test_data['id'] = test_dataset['id']
test_data['title'] = test_dataset['title']
test_data['context'] = test_dataset['context']
test_data['question'] = test_dataset['question']
test_data['question_type'] = test_dataset['question_type']
test_data['answers'] = test_dataset['answers']
test_data["context_length"] = 0
```

```python
for i in range(len(test_data.context)):
    test_data["context_length"][i] = len(test_data["context"][i])
```

```python
test_data.head()
```

```python
test_data.sample(5)
```

```python
test_document = test_data[["title", "context"]].drop_duplicates(subset="context").reset_index(drop=True)
test_document
```

```python
from helpers.utils import Preprocessing

preprocess = Preprocessing()
```

```python
stopwords_list = preprocess.get_stopwords("C:\\Python\\Pytorch\\Transformer related\\Project JaQUAd QA System\\Stopwords\\stopwords-ja.txt")

stopwords_list[-1]
```

```python
test_document["tokenized_context"] = test_document["context"].apply(lambda x: preprocess.get_tokenized_sentence(x, stopwords_list))

test_document.head()
```

```python
test_document["tokenized_context"][0]
```

```python
test_document= test_document.reset_index().rename(columns={"index":"doc_label"})
```

```python
test_document
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# defining the TF-IDF
tfidf_configs = {
    'token_pattern': r"(?u)\b\w\w+\b",
    'analyzer': 'word',
    'binary': True
}

# defining the number of documents to retrieve
retriever_configs = {
    'n_neighbors': 5,
    'metric': 'cosine'
}

embedding = TfidfVectorizer(**tfidf_configs)
retriever = NearestNeighbors(**retriever_configs)
```

```python
X = embedding.fit_transform(test_document['tokenized_context'])
retriever.fit(X, test_document['doc_label'])
```

```python
test_data["tokenized_question"] = test_data["question"].apply(lambda x: preprocess.get_tokenized_sentence(x, stopwords_list))

test_data.head()
```

```python
questions_embedding = embedding.transform(test_data["tokenized_question"])
q_pred = retriever.kneighbors(questions_embedding, return_distance=False)
q_pred[:10]
```

```python
test_data["doc_label"] = 0
for i in range(len(test_data.index)):
    for j in range(len(test_document.index)):
        if test_data["context"][i] == test_document["context"][j]:
            test_data["doc_label"][i] = test_document["doc_label"][j]
            break
```

```python
test_data["tokenized_context"] = test_data["context"].apply(lambda x: preprocess.get_tokenized_sentence(x, stopwords_list))
test_data.head()
```

```python
from helpers.utils import tools

tool = tools()

q_test = test_data["doc_label"]
acc = tool.top_accuracy(q_test, q_pred)
print('Accuracy:', f'{acc:.4f}')
print('Quantity:', int(acc*len(q_pred)), 'from', len(q_pred))
```
